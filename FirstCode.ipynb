## Pay Check Calculator

```python
print("Pay Check Calculator")
```

```python
taxrate = 18
```

```python
hours = int(input("Hours Worked: "))
```

```python
PayRate = float(input("Hourly Pay Rate: "))
```

```python
GrossPay = float(hours * PayRate)
```

```python
tax = GrossPay * (taxrate / 100)
```

```python
TakeHome = GrossPay - tax
```

```python
print("Gross Pay: " + str(GrossPay))
print("Tax Rate: " + str(taxrate) + "%")
print("Tax Amount: " + str(tax))
print("Take Home Pay: " + str(TakeHome))
```

## Travel Time Calculator

```python
print("Travel Time Calculator")
```

```python
miles = int(input("Enter miles: "))
```

```python
mph = int(input("Enter miles per hour: "))
```

```python
hr = miles // mph
```

```python
minutes = miles % mph
```

```python
print("Estimated travel time")
```

```python
print("Hours: " + str(hr))
print("Minutes: " + str(minutes))
```
